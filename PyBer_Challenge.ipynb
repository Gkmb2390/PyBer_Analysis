# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_by_city = pyber_data_df.groupby(['type']).count()['ride_id']
rides_by_city
```

```python
# 2. Get the total drivers for each city type
drivers_by_city = city_data_df.groupby(['type']).sum()['driver_count']
drivers_by_city
```

```python
#  3. Get the total amount of fares for each city type
fares_by_city = pyber_data_df.groupby(['type']).sum()['fare']
fares_by_city
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = fares_by_city / rides_by_city
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = fares_by_city / drivers_by_city
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(
    {"Total Rides": rides_by_city, 
    'Total Drivers': drivers_by_city, 
    'Total Fares': fares_by_city,
    'Average Fare per Ride': avg_fare_per_ride,
    'Average Fare per Driver': avg_fare_per_driver})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the 
#fares for each date where the indices are the city type and date.
fares_by_week_by_city = pyber_data_df.groupby(['date','type']).sum()[['fare']]
fares_by_week_by_city
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is 
#needed to use the 'pivot()' function. df = df.reset_index()
fare_week_city_pivot_df = fares_by_week_by_city.reset_index()
fare_week_city_pivot_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the 
#columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_week_pivot =fare_week_city_pivot_df.pivot_table(index=['date'],columns = ['type'], values=['fare'])
fare_week_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
# new_fare_week_pivot_df = fare_week_pivot.loc[(fare_week_pivot['date'] >= '2019-01-01') & (fare_week_pivot['date'] <= '2019-04-29')]
# new_fare_week_pivot_df
new_fare_week_pivot_df = fare_week_pivot.loc[
    (fare_week_pivot.index >= '2019-01-01') & 
    (fare_week_pivot.index <= '2019-04-29')]
#fare_week_pivot.index
new_fare_week_pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
new_fare_week_pivot_df.index = pd.to_datetime(new_fare_week_pivot_df.index)
new_fare_week_pivot_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_fare_week_pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
new_fare_week_resample = new_fare_week_pivot_df.resample('W').sum()
new_fare_week_resample
```

```python
# 9. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
# #Object Oriented Code Beginning
fig, ax=plt.subplots(figsize=(10,3))
ax.plot(new_fare_week_resample, label=['Rural', 'Suburban', 'Urban'])
#Create y label
ax.set_ylabel("Fare($USD)")
#Create a Title
ax.set_title("Total Fare by City Type")
#Add a legend
ax.legend()

#Show the Plot
plt.show()

#Save the Image to folder
#plt.savefig("analysis/PyBer_fare_summary.png")
```
